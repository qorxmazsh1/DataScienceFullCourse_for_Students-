```python
# Tuple creation
t1 = (1, 2, 3)
t2 = 1, 2, 3        # Parentheses optional!
t3 = tuple([1,2,3])
t4 = ()              # Empty tuple
t5 = (1,)            # Single element - COMMA is required!
t6 = (1)             # This is NOT a tuple - it's just int 1!

print(type(t5))  # <class 'tuple'>
print(type(t6))  # <class 'int'> !!
```

```python
# Tuple is IMMUTABLE
t = (1, 2, 3)
# t[0] = 10    # TypeError!

# BUT: mutable elements INSIDE tuple CAN be modified!
t = ([1, 2], [3, 4])
t[0].append(5)      # This works!
print(t)             # ([1, 2, 5], [3, 4])
```

```python
# CLASSIC TRAP: t[1] += [4]
t = (1, [2, 3])
try:
    t[1] += [4]   # Raises TypeError!
except TypeError as e:
    print(f"Error: {e}")
print(t)  # (1, [2, 3, 4]) - list STILL changed despite error!
# Why? += first does extend (succeeds) then tries to assign back (fails)
```

```python
# Tuple methods (only 2!)
t = (1, 2, 3, 2, 1)
print(t.count(2))   # 2
print(t.index(3))   # 2
```

```python
# Tuple unpacking
a, b, c = (1, 2, 3)
print(a, b, c)  # 1 2 3

# Extended unpacking
a, *b, c = (1, 2, 3, 4, 5)
print(a, b, c)  # 1 [2, 3, 4] 5

# Swap trick
a, b = 1, 2
a, b = b, a
print(a, b)  # 2 1
```

```python
# Named tuples
from collections import namedtuple
Point = namedtuple('Point', ['x', 'y'])
p = Point(3, 4)
print(p.x, p.y)       # 3 4
print(p[0], p[1])     # 3 4
print(p._asdict())    # {'x': 3, 'y': 4}
print(p._replace(x=10))  # Point(x=10, y=4)
print(p._fields)      # ('x', 'y')
```

```python
# Tuple vs List
# Tuple: hashable (can be dict key), less memory, faster
# List: mutable, more flexible

d = {(1, 2): "point"}   # Tuple as dict key - OK!
# d = {[1, 2]: "point"} # TypeError! List is not hashable

# No tuple comprehension!
gen = (x for x in range(5))  # This is a GENERATOR, not tuple!
print(type(gen))  # <class 'generator'>
# Use tuple() to convert: tuple(x for x in range(5))
```

```python
# hash() with tuples
print(hash((1, 2, 3)))        # Some hash value
# print(hash((1, 2, [3])))    # TypeError! Contains unhashable list
```

# 🔑 KEY TAKEAWAYS
---
1. **`(1)` is int, `(1,)` is tuple** — comma is essential for single-element tuple
2. **Tuples are immutable** but mutable elements inside CAN change
3. **`t[1] += [4]` raises error BUT still modifies the list** — classic trap!
4. **Only 2 methods**: `.count()` and `.index()`
5. **Tuples are hashable** (if all elements are hashable) — can be dict keys
6. **`(x for x in range(5))` is a generator**, not a tuple comprehension
7. **`*args` in functions collects as tuple**, not list
