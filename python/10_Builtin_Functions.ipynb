```python
# Math functions
print(abs(-5))        # 5
print(round(2.5))     # 2! (Banker's rounding!)
print(round(1.5))     # 2
print(round(3.5))     # 4! (rounds to nearest EVEN)
print(round(2.675, 2))  # 2.67 (not 2.68! float precision)
print(min(3, 1, 4))   # 1
print(max(3, 1, 4))   # 4
print(sum([1,2,3]))    # 6
print(pow(2, 10))      # 1024
print(pow(2, 10, 100)) # 24 (2**10 % 100)
print(divmod(17, 5))   # (3, 2)
```

```python
# round() BANKER'S ROUNDING - EXAM TRAP!
# When exactly halfway, rounds to nearest EVEN number
print(round(0.5))   # 0 (not 1!)
print(round(1.5))   # 2
print(round(2.5))   # 2 (not 3!)
print(round(3.5))   # 4
print(round(4.5))   # 4 (not 5!)

# Float precision trap
print(round(2.675, 2))  # 2.67 (not 2.68!)
# Because 2.675 is actually 2.6749999... in float
```

```python
# all() and any()
print(all([True, True, True]))    # True
print(all([True, False, True]))   # False
print(all([]))                     # True! (empty = vacuously true)

print(any([False, False, True]))  # True
print(any([False, False, False])) # False
print(any([]))                     # False

print(any([0, "", None, False, 0.0]))  # False (all falsy)
```

```python
# Iteration functions
# enumerate
for i, v in enumerate(["a", "b", "c"], start=1):
    print(f"{i}:{v}", end=" ")  # 1:a 2:b 3:c
print()

# zip - stops at shortest
print(list(zip([1,2,3], "ab")))  # [(1,'a'), (2,'b')]

# reversed
print(list(reversed([1,2,3])))  # [3, 2, 1]

# sorted
print(sorted("hello"))  # ['e', 'h', 'l', 'l', 'o'] - returns LIST!
print(sorted([3,1,2], reverse=True))  # [3, 2, 1]
```

```python
# zip advanced
# Create dict from two lists
keys = ["a", "b", "c"]
values = [1, 2, 3]
d = dict(zip(keys, values))
print(d)  # {'a': 1, 'b': 2, 'c': 3}

# Unzip
pairs = [(1, 'a'), (2, 'b'), (3, 'c')]
nums, chars = zip(*pairs)
print(nums)   # (1, 2, 3)
print(chars)  # ('a', 'b', 'c')
```

```python
# Object functions
print(id(42))              # Unique identity
print(type([]))            # <class 'list'>
print(isinstance(True, int))  # True
print(hash("hello"))       # Some hash value
print(callable(print))     # True
print(callable(42))        # False
print(dir([]))             # All list attributes
```

```python
# print() parameters
print("a", "b", "c", sep="-")      # a-b-c
print("hello", end="!\n")          # hello!
# print("text", file=open("out.txt","w"))  # Write to file
```

# 🔑 KEY TAKEAWAYS
---
1. **`round()` uses Banker's rounding** — `round(0.5) = 0`, `round(1.5) = 2`, `round(2.5) = 2`
2. **`round(2.675, 2) = 2.67`** not 2.68 (float precision!)
3. **`all([]) = True`** (vacuously true), **`any([]) = False`**
4. **`sorted()` returns a LIST** even for strings
5. **`zip()` stops at shortest** iterable
6. **`callable()` checks if object can be called** like a function
