```python
# Import types
import math                    # Import entire module
from math import sqrt, pi      # Import specific names
from math import *             # Import all (not recommended!)
import math as m               # Alias

print(math.pi)
print(sqrt(16))
print(m.e)
```

```python
# __name__ == "__main__"
# When a file is run directly: __name__ == "__main__"
# When a file is imported: __name__ == module name

print(f"This module's name: {__name__}")

# Common pattern:
# if __name__ == "__main__":
#     main()  # Only runs when file is executed directly
```

```python
# Module search path
import sys
# Python searches for modules in this order:
# 1. Current directory
# 2. PYTHONPATH environment variable
# 3. Standard library
# 4. Site-packages (pip installed)
print(sys.path[:3])  # First 3 search paths
```

```python
# __all__ - controls what 'from module import *' exports
# In a module:
# __all__ = ['func1', 'func2']  # Only these are exported with *

# dir() shows what's in a module
import json
print([x for x in dir(json) if not x.startswith('_')])
```

# 🔑 KEY TAKEAWAYS
---
1. **`__name__ == "__main__"`** — only True when file is run directly
2. **`__all__`** controls `from module import *`
3. **Avoid `from X import *`** — pollutes namespace
4. **Module search**: current dir → PYTHONPATH → stdlib → site-packages
5. **`__init__.py`** makes a directory a package
