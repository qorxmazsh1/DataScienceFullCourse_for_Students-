## 1.1 Built-in Data Types Overview
Python has several built-in data types. Every value in Python is an object and has a type.

```python

# Numeric Types
x_int = 42                    # int - integer, unlimited precision
x_float = 3.14                # float - floating point (double precision)
x_complex = 3 + 4j            # complex - complex number (real + imaginary)
x_bool = True                  # bool - boolean (subclass of int!)

# Text Type
x_str = "Hello, EPAM!"        # str - string (immutable Unicode sequence)

# Sequence Types
x_list = [1, 2, 3]            # list - mutable ordered sequence
x_tuple = (1, 2, 3)           # tuple - immutable ordered sequence
x_range = range(10)            # range - immutable sequence of numbers

# Mapping Type
x_dict = {"a": 1, "b": 2}     # dict - mutable key-value mapping

# Set Types
x_set = {1, 2, 3}             # set - mutable unordered, no duplicates
x_frozenset = frozenset([1,2]) # frozenset - immutable set

# Binary Types
x_bytes = b"hello"             # bytes - immutable byte sequence
x_bytearray = bytearray(5)    # bytearray - mutable byte sequence
x_memoryview = memoryview(b"hello")  # memoryview

# None Type
x_none = None                  # NoneType - absence of value

print("All built-in types demonstrated!")
print(x_bytes)
```

```python
# Check the type of each value
values = [42, 3.14, 3+4j, True, "hello", [1,2], (1,2), range(5),
          {"a":1}, {1,2}, frozenset([1,2]), b"hi", bytearray(3), None]

for val in values:
    print(f"{str(val):25s} -> type: {type(val).__name__}")
```

## 1.2 Type Checking: `type()`, `isinstance()`, `issubclass()`

```python
# type() - returns the exact type of an object
print(type(42))           # <class 'int'>
print(type(3.14))         # <class 'float'>
print(type("hello"))      # <class 'str'>
print(type(True))         # <class 'bool'>
print(type(None))         # <class 'NoneType'>
print(type([1, 2]))       # <class 'list'>
print(type((1, 2)))       # <class 'tuple'>
print(type({"a": 1}))     # <class 'dict'>
print(type({1, 2}))       # <class 'set'>
```

```python
# isinstance() - checks inheritance chain (more flexible than type!)
print(isinstance(42, int))          # True
print(isinstance(True, int))        # True! bool is subclass of int!
print(isinstance(True, bool))       # True
print(isinstance(42, bool))         # False - int is NOT a bool

# Check against multiple types
print(isinstance(42, (int, float)))    # True
print(isinstance(3.14, (int, float)))  # True
print(isinstance("hi", (int, float)))  # False
```

```python
# issubclass() - checks class inheritance relationships
print(issubclass(bool, int))    # True! bool inherits from int
print(issubclass(int, object))  # True - everything inherits from object
print(issubclass(int, bool))    # False - int does NOT inherit from bool
```

```python
# TRAP: type() vs isinstance()
x = True
print(type(x) == int)       # False - type is exactly bool, not int
print(type(x) == bool)      # True  - type is exactly bool
print(isinstance(x, int))   # True  - bool IS an int (subclass)
```

## 1.3 Type Conversion / Casting

```python
# int() conversions
print(int(3.9))          # 3 - truncates (does NOT round!)
print(int(-3.9))         # -3 - truncates toward zero
print(int("42"))         # 42 - string to int
print(int("0b1010", 2)) # 10 - binary string
print(int("0xff", 16))  # 255 - hex string
print(int(True))         # 1
print(int(False))        # 0
# int("3.14")   # ValueError! Can't convert float string directly
# int("")       # ValueError! Empty string
```

```python
# float() conversions
print(float(42))        # 42.0
print(float("3.14"))    # 3.14
print(float("inf"))     # inf - positive infinity
print(float("-inf"))    # -inf - negative infinity
print(float("nan"))     # nan - Not a Number
print(float(True))      # 1.0
print(float(False))     # 0.0
```

```python
# str() conversions
print(str(42))          # '42'
print(str(3.14))        # '3.14'
print(str(True))        # 'True' (NOT '1'!)
print(str(None))        # 'None'
print(str([1, 2, 3]))   # '[1, 2, 3]'
```

```python
# bool() - Truthy and Falsy
print("=== FALSY VALUES ===")
print(bool(0))          # False
print(bool(0.0))        # False
print(bool(0j))         # False
print(bool(""))          # False - empty string
print(bool([]))          # False - empty list
print(bool(()))          # False - empty tuple
print(bool({}))          # False - empty dict
print(bool(set()))       # False - empty set
print(bool(None))        # False
print(bool(False))       # False
print(bool(range(0)))    # False - empty range
```

```python
# TRUTHY values - WATCH OUT FOR TRAPS!
print("=== TRUTHY VALUES ===")
print(bool(1))           # True - any nonzero number
print(bool(-1))          # True
print(bool(" "))         # True! TRAP: space is NOT empty!
print(bool("False"))     # True! TRAP: string "False" is truthy!
print(bool("0"))         # True! TRAP: string "0" is truthy!
print(bool([0]))         # True - non-empty list (even if element is falsy)
print(bool([False]))     # True - non-empty list
print(bool({0: False}))  # True - non-empty dict
```

## 1.5 ⭐ Mutable vs Immutable (MOST IMPORTANT TOPIC!)

```python
# === IMMUTABLE TYPES ===
# int, float, str, tuple, frozenset, bytes, bool
# Cannot be changed after creation

# String is immutable
s = "hello"
# s[0] = "H"   # TypeError: 'str' does not support item assignment
s_new = "H" + s[1:]
print(s_new)  # "Hello"
print(s)      # "hello" - original unchanged

# Tuple is immutable
t = (1, 2, 3)
# t[0] = 10   # TypeError

# int is immutable
a = 5
b = a
a = a + 1     # Creates NEW int, doesn't modify old one
print(a)      # 6
print(b)      # 5 - unchanged!
```

```python
# === MUTABLE TYPES ===
# list, dict, set, bytearray
# CAN be changed in-place

lst = [1, 2, 3]
lst[0] = 10
print(lst)         # [10, 2, 3]

d = {"a": 1}
d["b"] = 2
print(d)           # {'a': 1, 'b': 2}

s = {1, 2, 3}
s.add(4)
print(s)           # {1, 2, 3, 4}
```

```python
# CRITICAL: Aliasing with mutable objects
a = [1, 2, 3]
b = a           # b is NOT a copy - same list!
b.append(4)
print(a)        # [1, 2, 3, 4] - a is ALSO modified!
print(b)        # [1, 2, 3, 4]
print(a is b)   # True - SAME object
```

```python
# Aliasing does NOT cause issues with immutable types
a = "hello"
b = a
a = a + " world"  # Creates NEW string
print(a)  # "hello world"
print(b)  # "hello" - unchanged!

x = 10
y = x
x += 1    # Creates NEW int (11)
print(x)  # 11
print(y)  # 10 - unchanged!
```

```python
# Mutable objects in functions - changes PERSIST!
def modify_list(lst):
    lst.append(999)

my_list = [1, 2, 3]
modify_list(my_list)
print(my_list)  # [1, 2, 3, 999] - modified!

def modify_string(s):
    s = s + " world"  # Local rebind only

my_str = "hello"
modify_string(my_str)
print(my_str)   # "hello" - unchanged!
```

## 1.6 Variable Assignment — Reference Semantics

```python
# Variables are REFERENCES to objects
a = [1, 2, 3]
b = a
print(id(a) == id(b))  # True - same object
print(a is b)           # True
```

```python
# Multiple assignment
a, b, c = 1, 2, 3
print(a, b, c)  # 1 2 3

# Swap values
a, b = b, a
print(a, b)     # 2 1

# Same value to multiple variables
x = y = z = 0
print(x, y, z)  # 0 0 0
```

```python
# TRAP: Multiple assignment with MUTABLE objects
a = b = []       # Both point to SAME list!
a.append(1)
print(b)          # [1] - b is affected!

# Correct way - separate objects:
a = []
b = []
a.append(1)
print(b)          # [] - independent!
```

## 1.7 `id()` Function — Object Identity

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]  # Different object, same content

print(f"id(a) = {id(a)}")
print(f"id(b) = {id(b)}")
print(f"id(c) = {id(c)}")
print(f"a is b: {a is b}")  # True - same object
print(f"a is c: {a is c}")  # False - different objects
print(f"a == c: {a == c}")  # True - same VALUE
```

## 1.8 Integer Caching / Interning

```python
# CPython caches integers in range [-5, 256]
a = 256
b = 256
print(a is b)  # True - cached

a = -5
b = -5
print(a is b)  # True - in cache range

# Outside cache range:
a = 257
b = 257
print(a == b)  # True - always
# a is b  # Implementation-dependent!
```

## 1.9 String Interning

```python
# Python auto-interns some strings (identifiers)
a = "hello"
b = "hello"
print(a is b)  # True - interned

# Manual interning
import sys
a = sys.intern("hello world")
b = sys.intern("hello world")
print(a is b)  # True

# ALWAYS use == for string comparison, not 'is'!
```

## 1.15 Float Precision Issues (CLASSIC EXAM QUESTION!)

```python
# THE CLASSIC TRAP: 0.1 + 0.2 != 0.3
print(0.1 + 0.2)           # 0.30000000000000004
print(0.1 + 0.2 == 0.3)    # False!

# Solution 1: math.isclose()
import math
print(math.isclose(0.1 + 0.2, 0.3))  # True

# Solution 2: decimal module
from decimal import Decimal
print(Decimal('0.1') + Decimal('0.2') == Decimal('0.3'))  # True
# Use STRING input for Decimal, not float!
```

## 1.16 `bool` is a Subclass of `int`

```python
# bool inherits from int!
print(True + True)          # 2
print(True + True + True)   # 3
print(True * 10)            # 10
print(type(True + True))    # <class 'int'> - result is int!

print(isinstance(True, int))  # True
print(issubclass(bool, int))  # True

# Dict key trap!
d = {True: "yes", 1: "one", 1.0: "one_float"}
print(d)  # {True: 'one_float'} - all same key!
```

```python
# GOTCHA 1: is vs ==
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)   # True - same value
print(a is b)   # False - different objects!

c = a
print(a is c)   # True - same object
```

```python
# GOTCHA 2: Empty containers are falsy but != False
print([] == False)        # False!
print(bool([]) == False)  # True
print("" == False)        # False!
print(0 == False)         # True (bool subclass of int)
print(1 == True)          # True
print(2 == True)          # False - only 1 equals True
```

```python
# GOTCHA 3: += vs + with mutable/immutable
# List (mutable): += modifies in place
a = [1, 2]
old_id = id(a)
a += [3]
print(id(a) == old_id)  # True - same object!

# Tuple (immutable): += creates new object
b = (1, 2)
old_id = id(b)
b += (3,)
print(id(b) == old_id)  # False - new object!
```

```python
# GOTCHA 4: [] is [] is False
print([] is [])   # False - different objects
print([] == [])   # True - same value
print(() is ())   # True! Empty tuple is cached
```

```python
# GOTCHA 5: Boolean arithmetic types
print(type(True + True))    # int
print(type(True + 0.0))     # float
print(type(True + 0j))      # complex

# Boolean indexing
lst = ["zero", "one"]
print(lst[True])   # "one"
print(lst[False])  # "zero" 
```

```python
# GOTCHA 6: String comparison is lexicographic
print("9" > "10")      # True! '9' > '1'
print("100" < "2")     # True! '1' < '2'
```

```python
# GOTCHA 7: NaN != NaN
nan = float('nan')
print(nan == nan)      # False!
print(nan is nan)      # True (same object)
import math
print(math.isnan(nan)) # True - correct way to check
```

```python
# GOTCHA 8: Division always returns float
print(type(10 / 2))    # float!
print(10 / 2)          # 5.0 not 5!
print(type(10 // 2))   # int
print(10 // 2)         # 5
```

# 🔑 KEY TAKEAWAYS
---

1. **`bool` is a subclass of `int`** — `True == 1`, `False == 0`, `isinstance(True, int)` is `True`

2. **Mutable vs Immutable** — Lists, dicts, sets are mutable; ints, strings, tuples are immutable. This affects aliasing!

3. **`is` vs `==`** — `is` checks identity (same object), `==` checks equality (same value). Use `is` with `None`.

4. **Integer caching** — CPython caches [-5, 256]. `a = 256; b = 256; a is b` → `True`

5. **`+=` vs `+`** — For lists, `+=` modifies in-place (extend), `+` creates new list. Critical difference!

6. **Falsy values** — `0, 0.0, "", [], (), {}, set(), None, False`. Everything else is truthy.

7. **`bool(" ")` is `True`** — String with space is NOT empty! Same for `bool("0")`, `bool("False")`

8. **Float precision** — `0.1 + 0.2 != 0.3`. Use `math.isclose()` or `Decimal`.

9. **`int()` truncates, doesn't round** — `int(3.9)` → `3`, `int(-3.9)` → `-3`

10. **`{True: "a", 1: "b", 1.0: "c"}` has only 1 key** — Because `True == 1 == 1.0`
