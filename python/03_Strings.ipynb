## 3.1 String Creation

```python
# String creation methods
s1 = 'single quotes'
s2 = "double quotes"
s3 = """triple quotes
can span multiple lines"""
s4 = r"raw string: \n is NOT a newline"

print(s1)
print(s2)
print(s3)
print(s4)  # prints \n literally
```

## 3.2 String Immutability

```python
s = "hello"
# s[0] = "H"   # TypeError! Strings are immutable!
s_new = "H" + s[1:]
print(s_new)  # "Hello"
print(s)      # "hello" - unchanged
```

## 3.3 Indexing & Slicing

```python
s = "Hello, World!"
print(s[0])      # 'H'
print(s[-1])     # '!' 
print(s[0:5])    # 'Hello'
print(s[7:])     # 'World!'
print(s[:5])     # 'Hello'
print(s[::2])    # 'Hlo ol!'  every 2nd char
print(s[::-1])   # '!dlroW ,olleH'  reversed

# TRAP: index out of range = Error, but slice out of range = no error!
# print(s[100])  # IndexError!
print(s[1:100])  # 'ello, World!' - no error!
print(s[100:200]) # '' - empty string, no error!
```

## 3.4 String Methods

```python
s = "Hello, World!"

# Case methods
print(s.upper())       # 'HELLO, WORLD!'
print(s.lower())       # 'hello, world!'
print(s.title())       # 'Hello, World!'
print(s.capitalize())  # 'Hello, world!'
print(s.swapcase())    # 'hELLO, wORLD!'
print("straße".casefold())  # 'strasse' - aggressive lowercase
```

```python
# Search methods
s = "hello world hello"
print(s.find("world"))     # 6 (index)
print(s.find("xyz"))       # -1 (not found)
# print(s.index("xyz"))    # ValueError! (index raises error)
print(s.count("hello"))    # 2
print(s.rfind("hello"))    # 12 (last occurrence)

# find() returns -1, index() raises ValueError - KEY DIFFERENCE!
```

```python
# Validation methods
print("abc".isalpha())     # True
print("123".isdigit())     # True
print("abc123".isalnum())  # True
print("   ".isspace())     # True
print("Hello World".istitle())  # True
print("HELLO".isupper())   # True
print("hello".islower())   # True
print("var_name".isidentifier())  # True
```

```python
# Strip, replace
s = "  hello  "
print(s.strip())    # 'hello'
print(s.lstrip())   # 'hello  '
print(s.rstrip())   # '  hello'

print("hello world".replace("world", "EPAM"))  # 'hello EPAM'
print("aabbcc".replace("b", "x", 1))  # 'aaxbcc' (replace 1 occurrence)
```

```python
# Split and join
print("a,b,c".split(","))           # ['a', 'b', 'c']
print("hello   world".split())      # ['hello', 'world'] - any whitespace
print("hello   world".split(" "))   # ['hello', '', '', 'world'] - exact split!
print(",".join(["a", "b", "c"]))    # 'a,b,c'
print(" ".join(["Hello", "EPAM"]))  # 'Hello EPAM'

# TRAP: split() vs split(" ") are DIFFERENT!
# split() collapses whitespace, split(" ") does NOT
```

```python
# Alignment
print("hello".center(20, "-"))  # '-------hello--------'
print("hello".ljust(20, "."))   # 'hello...............'
print("hello".rjust(20, "."))   # '...............hello'
print("42".zfill(8))            # '00000042'

# startswith / endswith
print("hello".startswith("hel"))   # True
print("hello".endswith(("lo", "la")))  # True (tuple of suffixes)
```

## 3.5 String Formatting

```python
name, age = "Alice", 25

# Method 1: % formatting (old style)
print("Name: %s, Age: %d" % (name, age))

# Method 2: .format()
print("Name: {}, Age: {}".format(name, age))
print("Name: {n}, Age: {a}".format(n=name, a=age))

# Method 3: f-strings (Python 3.6+) - PREFERRED!
print(f"Name: {name}, Age: {age}")
print(f"{3.14159:.2f}")     # '3.14' - 2 decimal places
print(f"{'hello':>20}")     # '               hello' - right-aligned
print(f"{42:08b}")          # '00101010' - binary, zero-padded
print(f"{name=}")           # "name='Alice'" (Python 3.8+ debugging)
```

## 3.6 Escape Sequences & ord()/chr()

```python
print("Hello\nWorld")    # newline
print("Hello\tWorld")    # tab
print("Hello\\World")    # literal backslash

# ord() and chr()
print(ord("A"))    # 65
print(ord("a"))    # 97
print(chr(65))     # A
print(chr(97))     # a
```

## 3.7 String Concatenation Performance

```python
# + creates new string each time (slow for many strings)
# .join() is much faster for joining many strings!

words = ["hello"] * 1000

# Slow way:
# result = ""
# for w in words: result += w

# Fast way:
result = "".join(words)
print(len(result))  # 5000
```

## 3.8 Bytes vs Strings

```python
# str = text (Unicode), bytes = binary data
s = "hello"
b = b"hello"
print(type(s))  # <class 'str'>
print(type(b))  # <class 'bytes'>

# Encoding: str -> bytes
encoded = s.encode('utf-8')
print(encoded)  # b'hello'

# Decoding: bytes -> str
decoded = encoded.decode('utf-8')
print(decoded)  # 'hello' 
```

```python
# TRAP: "abc" > "abd" - lexicographic comparison
print("abc" < "abd")   # True ('c' < 'd')
print("abc" < "abcd")  # True (shorter is "less")
print("9" > "10")      # True! ('9' > '1')
```

```python
# TRAP: split() vs split(" ")
print("a  b  c".split())      # ['a', 'b', 'c']
print("a  b  c".split(" "))   # ['a', '', 'b', '', 'c']
```

```python
# TRAP: isspace(), isdigit() etc.
print("   ".isspace())   # True (only whitespace)
print("".isspace())      # False! (empty string returns False for all is* methods)
print("123".isdigit())   # True
print("12.3".isdigit())  # False (has a dot)
```

```python
# TRAP: "" == False
print("" == False)        # False!
print(bool("") == False)  # True
print(bool("") is False)  # True
```

# 🔑 KEY TAKEAWAYS
---
1. Strings are **immutable** — can't change in place
2. **Slicing** beyond bounds is OK (no error), **indexing** beyond bounds raises Error
3. **`find()` returns -1**, **`index()` raises ValueError** when not found
4. **`split()` vs `split(" ")`** — split() collapses whitespace, split(" ") doesn't
5. **f-strings** are preferred for formatting (Python 3.6+)
6. **Empty string is falsy** but `"" == False` is `False`
7. **`is*` methods return False on empty strings**
8. Use **`.join()`** instead of `+` for concatenating many strings (faster)
