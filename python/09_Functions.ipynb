```python
# Function basics
def greet(name):
    return f"Hello, {name}!"

print(greet("EPAM"))  # "Hello, EPAM!"

# No return = returns None
def no_return():
    x = 42

print(no_return())  # None

# Multiple return values (actually returns tuple)
def min_max(lst):
    return min(lst), max(lst)

result = min_max([3, 1, 4, 1, 5])
print(result)        # (1, 5)
print(type(result))  # <class 'tuple'>
```

## Argument Types

```python
# Positional and keyword arguments
def func(a, b, c):
    print(a, b, c)

func(1, 2, 3)         # Positional
func(a=1, b=2, c=3)   # Keyword
func(1, c=3, b=2)     # Mixed (positional first!)
```

```python
# Default arguments
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Alice"))           # "Hello, Alice!"
print(greet("Bob", "Hi"))       # "Hi, Bob!" 
```

```python
# MUTABLE DEFAULT ARGUMENT TRAP (MOST CLASSIC QUESTION!)
def append_to(element, lst=[]):
    lst.append(element)
    return lst

print(append_to(1))   # [1]
print(append_to(2))   # [1, 2] !! Default list is shared!
print(append_to(3))   # [1, 2, 3] !! Keeps growing!

# CORRECT WAY:
def append_to_fixed(element, lst=None):
    if lst is None:
        lst = []
    lst.append(element)
    return lst

print(append_to_fixed(1))  # [1]
print(append_to_fixed(2))  # [2] - independent!
```

```python
# *args and **kwargs
def func(*args, **kwargs):
    print(f"args: {args}")      # tuple
    print(f"kwargs: {kwargs}")  # dict

func(1, 2, 3, name="Alice", age=25)
# args: (1, 2, 3)
# kwargs: {'name': 'Alice', 'age': 25}
```

```python
# Keyword-only arguments (after *)
def func(a, *, b, c):
    print(a, b, c)

func(1, b=2, c=3)     # OK
# func(1, 2, 3)        # TypeError! b and c must be keyword

# Positional-only arguments (before /, Python 3.8+)
def func(a, b, /, c, d):
    print(a, b, c, d)

func(1, 2, c=3, d=4)   # OK
# func(a=1, b=2, c=3, d=4)  # TypeError! a,b must be positional

# Full syntax
def func(pos_only, /, normal, *, kw_only):
    pass
```

## LEGB Scope Rule

```python
# L - Local, E - Enclosing, G - Global, B - Built-in
x = "global"

def outer():
    x = "enclosing"
    def inner():
        x = "local"
        print(x)    # "local" (L)
    inner()
    print(x)        # "enclosing" (E)

outer()
print(x)            # "global" (G)
print(len)          # <built-in function len> (B)
```

```python
# global keyword
x = 10
def modify_global():
    global x
    x = 20

modify_global()
print(x)  # 20

# nonlocal keyword
def outer():
    x = 10
    def inner():
        nonlocal x
        x = 20
    inner()
    print(x)  # 20

outer()
```

## Lambda Functions

```python
# Lambda: anonymous, single expression
square = lambda x: x ** 2
print(square(5))  # 25

# Common use: as key function
lst = [(1, 'b'), (2, 'a'), (3, 'c')]
print(sorted(lst, key=lambda x: x[1]))  # [(2,'a'), (1,'b'), (3,'c')]
```

## Higher-Order Functions

```python
# map, filter, reduce
nums = [1, 2, 3, 4, 5]

# map: apply function to each element
print(list(map(lambda x: x**2, nums)))    # [1, 4, 9, 16, 25]

# filter: keep elements where function returns True
print(list(filter(lambda x: x%2==0, nums)))  # [2, 4]

# reduce: combine elements pairwise
from functools import reduce
print(reduce(lambda a, b: a + b, nums))   # 15
```

## Type Hints

```python
# Type hints have NO runtime effect - just for docs/IDE
def greet(name: str, times: int = 1) -> str:
    return (f"Hello, {name}! ") * times

print(greet("EPAM", 2))

# The function works even with wrong types:
print(greet(42))  # No error! Type hints aren't enforced
```

# 🔑 KEY TAKEAWAYS
---
1. **MUTABLE DEFAULT ARGUMENT** is #1 trap! `def f(lst=[])` shares the list. Use `None` instead.
2. **`*args` is tuple, `**kwargs` is dict**
3. **LEGB Rule**: Local → Enclosing → Global → Built-in
4. **`global`** modifies global variable, **`nonlocal`** modifies enclosing scope
5. **Lambda** is single expression only, no statements
6. **`return a, b`** actually returns a tuple
7. **Function without return** returns `None`
8. **Type hints** are NOT enforced at runtime
9. **`/` for positional-only** (3.8+), **`*` for keyword-only** arguments
