```python
# File modes
# 'r'  - Read (default), file must exist
# 'w'  - Write, creates/truncates file
# 'a'  - Append, writes to end
# 'x'  - Exclusive create, error if exists
# 'r+' - Read+Write
# 'b'  - Binary mode (rb, wb, ab)

# ALWAYS use 'with' statement - auto-closes file!
with open("test_file.txt", "w") as f:
    f.write("Hello, EPAM!\n")
    f.write("Line 2\n")

with open("test_file.txt", "r") as f:
    content = f.read()
    print(content)
```

```python
# Reading methods
with open("test_file.txt", "r") as f:
    # read() - entire file as string
    content = f.read()
    print("read():", repr(content))

with open("test_file.txt", "r") as f:
    # readline() - one line at a time
    line1 = f.readline()
    line2 = f.readline()
    print("readline():", repr(line1), repr(line2))

with open("test_file.txt", "r") as f:
    # readlines() - list of lines
    lines = f.readlines()
    print("readlines():", lines)

with open("test_file.txt", "r") as f:
    # Iteration (most Pythonic, memory efficient)
    for line in f:
        print("iter:", line.strip())
```

```python
# seek() and tell()
with open("test_file.txt", "r") as f:
    print(f.tell())     # 0 (beginning)
    f.read(5)
    print(f.tell())     # 5 (moved forward)
    f.seek(0)           # Go back to beginning
    print(f.tell())     # 0
```

```python
# JSON handling
import json

data = {"name": "Alice", "age": 25, "scores": [90, 85, 92]}

# Write JSON
with open("test_data.json", "w") as f:
    json.dump(data, f, indent=2)

# Read JSON
with open("test_data.json", "r") as f:
    loaded = json.load(f)
print(loaded)

# String conversion
json_str = json.dumps(data)
data_back = json.loads(json_str)
print(json_str)
```

```python
# pathlib (modern approach)
from pathlib import Path

p = Path("test_file.txt")
print(p.exists())       # True
print(p.is_file())      # True
print(p.suffix)         # '.txt'
print(p.stem)           # 'test_file'
content = p.read_text()
print(content)

# Cleanup
import os
for f in ["test_file.txt", "test_data.json"]:
    if os.path.exists(f):
        os.remove(f)
```

# 🔑 KEY TAKEAWAYS
---
1. **Always use `with` statement** for file handling
2. **`'r'` is default mode**, `'w'` truncates existing files
3. **`json.dump`/`json.load`** for files, **`json.dumps`/`json.loads`** for strings
4. **`pathlib`** is the modern way to handle paths
5. **Iterate over file** (`for line in f`) for memory efficiency
