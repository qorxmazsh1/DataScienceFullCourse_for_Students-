```python
# if / elif / else
x = 15
if x > 20:
    print("big")
elif x > 10:
    print("medium")  # This prints
else:
    print("small")
```

```python
# Ternary expression
x = 10
result = "even" if x % 2 == 0 else "odd"
print(result)  # "even"

# Nested ternary (avoid for readability)
x = 0
result = "positive" if x > 0 else "zero" if x == 0 else "negative"
print(result)  # "zero" 
```

```python
# for loops
for i in range(5):
    print(i, end=" ")  # 0 1 2 3 4
print()

# range(start, stop, step) - stop is NOT included!
for i in range(2, 10, 3):
    print(i, end=" ")  # 2 5 8
print()

# Negative range
for i in range(5, 0, -1):
    print(i, end=" ")  # 5 4 3 2 1
print()

# enumerate
for i, char in enumerate("abc", start=1):
    print(f"{i}:{char}", end=" ")  # 1:a 2:b 3:c
print()

# zip
for a, b in zip([1,2,3], ["a","b","c"]):
    print(f"{a}-{b}", end=" ")  # 1-a 2-b 3-c
```

```python
# while loop
count = 0
while count < 5:
    print(count, end=" ")
    count += 1  # 0 1 2 3 4
```

```python
# break, continue, pass
for i in range(10):
    if i == 3:
        continue   # Skip 3
    if i == 7:
        break      # Stop at 7
    print(i, end=" ")  # 0 1 2 4 5 6
```

## for...else and while...else (VERY IMPORTANT!)

```python
# else block runs when loop completes WITHOUT break
for i in range(5):
    if i == 10:  # Never true
        break
else:
    print("Loop completed without break!")  # This prints!

# With break:
for i in range(5):
    if i == 3:
        break
else:
    print("This won't print")  # NOT printed (break occurred)
```

```python
# TRAP: Empty loop still triggers else!
for i in range(0):  # Empty range
    print("never")
else:
    print("else runs!")  # This prints!

# while False also triggers else!
while False:
    print("never")
else:
    print("else runs!")  # This prints!
```

```python
# pass with else
for i in range(5):
    pass
else:
    print("done")  # Prints! pass doesn't trigger else bypass
```

## match-case (Python 3.10+)

```python
# Structural pattern matching
command = "quit"
match command:
    case "start":
        print("Starting...")
    case "quit" | "exit":
        print("Goodbye!")  # This matches
    case _:
        print("Unknown command")
```

```python
# TRAP: for...else with break is the key pattern
# Use case: searching for something
numbers = [1, 3, 5, 7, 9]
target = 5
for n in numbers:
    if n == target:
        print(f"Found {target}!")
        break
else:
    print(f"{target} not found")  # Only if break didn't happen
```

# 🔑 KEY TAKEAWAYS
---
1. **`for...else`**: else runs when loop completes **without break** — EXAM FAVORITE!
2. **Empty loops** (`range(0)`, `while False`) still trigger else!
3. **`break`** only exits the innermost loop
4. **`range(stop)`** does NOT include stop
5. **`zip()`** stops at shortest iterable
6. **Ternary**: `x if condition else y` — condition in the MIDDLE
7. **`match-case`** is Python 3.10+
