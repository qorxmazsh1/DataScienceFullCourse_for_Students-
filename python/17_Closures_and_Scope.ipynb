```python
# LEGB Rule detailed
x = "global"

def outer():
    x = "enclosing"
    def inner():
        x = "local"
        print(f"inner: {x}")  # local
    inner()
    print(f"outer: {x}")  # enclosing

outer()
print(f"global: {x}")  # global
```

```python
# Closure: inner function remembers enclosing scope
def make_counter():
    count = 0
    def counter():
        nonlocal count
        count += 1
        return count
    return counter

c = make_counter()
print(c())  # 1
print(c())  # 2
print(c())  # 3
```

```python
# __closure__ attribute
def make_multiplier(n):
    def multiply(x):
        return x * n
    return multiply

times3 = make_multiplier(3)
print(times3(10))  # 30
print(times3.__closure__)  # Shows closure cells
print(times3.__closure__[0].cell_contents)  # 3
```

```python
# LATE BINDING TRAP (VERY IMPORTANT!)
funcs = [lambda: i for i in range(5)]
results = [f() for f in funcs]
print(results)  # [4, 4, 4, 4, 4] - ALL 4!
# Why? Lambda captures variable 'i' by reference, not value
# When called, i is already 4 (loop finished)

# FIX: Use default argument to capture value at creation time
funcs = [lambda i=i: i for i in range(5)]
results = [f() for f in funcs]
print(results)  # [0, 1, 2, 3, 4] - correct!
```

```python
# nonlocal vs global
x = 10

def func():
    global x  # Modify global
    x = 20

func()
print(x)  # 20

def outer():
    x = 10
    def inner():
        nonlocal x  # Modify enclosing
        x = 20
    inner()
    print(x)  # 20

outer()
```

# 🔬 DEEP DIVE — Traps & Gotchas
---

```python
# DEEP DIVE: Late binding explained step by step
funcs = []
for i in range(3):
    funcs.append(lambda: i)
    # Each lambda captures REFERENCE to i, not VALUE

# After loop: i = 2
print(funcs[0]())  # 2 — i is 2 now!
print(funcs[1]())  # 2 — i is 2 now!
print(funcs[2]())  # 2 — i is 2 now!

# THREE FIXES:
# Fix 1: Default argument
funcs1 = [lambda i=i: i for i in range(3)]
print([f() for f in funcs1])  # [0, 1, 2]

# Fix 2: functools.partial
from functools import partial
funcs2 = [partial(lambda i: i, i) for i in range(3)]
print([f() for f in funcs2])  # [0, 1, 2]

# Fix 3: Closure factory
def make_func(i):
    return lambda: i
funcs3 = [make_func(i) for i in range(3)]
print([f() for f in funcs3])  # [0, 1, 2]
```

```python
# TRAP: UnboundLocalError
x = 10
def func():
    # Python sees x = ... later, so treats x as local
    print(x)  # UnboundLocalError! x assigned later
    x = 20

# Fix: use 'global x' or restructure
def func_fixed():
    global x
    print(x)
    x = 20
```

```python
# Solution 1
def make_accumulator(initial=0):
    total = initial
    def add(n):
        nonlocal total
        total += n
        return total
    return add

acc = make_accumulator()
assert acc(5) == 5
assert acc(3) == 8
assert acc(2) == 10

acc2 = make_accumulator(100)
assert acc2(10) == 110
print("\u2705 Exercise 1 passed!")
```

# 🔑 KEY TAKEAWAYS
---
1. **LEGB Rule**: Local → Enclosing → Global → Built-in
2. **Late binding trap**: `[lambda: i for i in range(5)]` → all return 4! Fix with `lambda i=i: i`
3. **`nonlocal`** modifies enclosing scope, **`global`** modifies module scope
4. **Closure** = inner function that remembers enclosing variables
5. **`__closure__`** attribute shows captured variables
