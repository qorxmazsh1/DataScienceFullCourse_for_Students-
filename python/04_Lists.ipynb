## List Methods

```python
lst = [1, 2, 3]
lst.append(4)         # [1, 2, 3, 4] - adds ONE element
lst.extend([5, 6])    # [1, 2, 3, 4, 5, 6] - adds MULTIPLE elements
lst.insert(0, 0)      # [0, 1, 2, 3, 4, 5, 6] - insert at index
print(lst)

# TRAP: append vs extend
a = [1, 2]
a.append([3, 4])    # [1, 2, [3, 4]] - adds LIST as single element!
print(a)

b = [1, 2]
b.extend([3, 4])    # [1, 2, 3, 4] - adds each element
print(b)
```

```python
lst = [3, 1, 4, 1, 5, 9]
lst.remove(1)       # Removes FIRST occurrence: [3, 4, 1, 5, 9]
print(lst)
popped = lst.pop()  # Removes & returns last: 9
print(popped, lst)
popped = lst.pop(0) # Removes & returns at index 0: 3
print(popped, lst)
print(lst.count(1))  # 1
print(lst.index(4))  # 0
```

```python
# TRAP: sort() returns None!
lst = [3, 1, 4, 1, 5]
result = lst.sort()
print(result)   # None! sort() modifies in-place
print(lst)      # [1, 1, 3, 4, 5]

# sorted() returns NEW list
lst2 = [3, 1, 4, 1, 5]
result = sorted(lst2)
print(result)   # [1, 1, 3, 4, 5]
print(lst2)     # [3, 1, 4, 1, 5] - unchanged!

# reverse() also returns None
lst.reverse()
print(lst)  # [5, 4, 3, 1, 1]
```

## List Comprehension

```python
# Basic
print([x for x in range(10)])                    # [0,1,...,9]

# With condition
print([x for x in range(10) if x % 2 == 0])     # [0,2,4,6,8]

# If-else (expression comes BEFORE for!)
print([x if x > 0 else -x for x in [-2,-1,0,1,2]])  # [2,1,0,1,2]

# Nested
matrix = [[1,2],[3,4],[5,6]]
flat = [x for row in matrix for x in row]
print(flat)  # [1,2,3,4,5,6]

# Multiple conditions
print([x for x in range(30) if x % 2 == 0 if x % 3 == 0])  # [0,6,12,18,24]
```

## Shallow Copy vs Deep Copy

```python
import copy

# Reference (NOT a copy)
a = [[1, 2], [3, 4]]
b = a           # Same object!
b[0].append(99)
print(a)        # [[1, 2, 99], [3, 4]] - affected!

# Shallow copy
a = [[1, 2], [3, 4]]
b = a.copy()    # or a[:] or list(a)
b[0].append(99)
print(a)        # [[1, 2, 99], [3, 4]] - STILL affected! Nested lists shared!
b.append([5, 6])
print(a)        # [[1, 2, 99], [3, 4]] - outer list NOT affected

# Deep copy
a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)
b[0].append(99)
print(a)        # [[1, 2], [3, 4]] - NOT affected! Fully independent
```

## List Multiplication Trap

```python
# TRAP: [[0]*3]*3
a = [[0] * 3] * 3
print(a)  # [[0, 0, 0], [0, 0, 0], [0, 0, 0]]
a[0][0] = 1
print(a)  # [[1, 0, 0], [1, 0, 0], [1, 0, 0]] - ALL rows changed!
# Because inner lists are the SAME reference!

# Correct way:
a = [[0] * 3 for _ in range(3)]
a[0][0] = 1
print(a)  # [[1, 0, 0], [0, 0, 0], [0, 0, 0]] - only first row changed
```

## List Unpacking

```python
a, b, *rest = [1, 2, 3, 4, 5]
print(a, b, rest)  # 1 2 [3, 4, 5]

first, *middle, last = [1, 2, 3, 4, 5]
print(first, middle, last)  # 1 [2, 3, 4] 5

# * unpacking in print
print(*[1, 2, 3])  # 1 2 3 (not [1, 2, 3])
```

## del Statement

```python
lst = [0, 1, 2, 3, 4, 5]
del lst[0]      # Remove by index
print(lst)      # [1, 2, 3, 4, 5]
del lst[1:3]    # Remove slice
print(lst)      # [1, 4, 5]
```

# 🔬 DEEP DIVE
---

```python
# += vs + for lists
a = [1,2,3]; b = a
b += [4]         # In-place extend
print(a)         # [1,2,3,4] - affected!

a = [1,2,3]; b = a
b = b + [4]      # New list
print(a)         # [1,2,3] - NOT affected!
```

# 🔑 KEY TAKEAWAYS
---
1. **`append()` vs `extend()`** — append adds one element, extend adds multiple
2. **`sort()` returns None** — modifies in-place. Use `sorted()` for new list
3. **`[[0]*3]*3` trap** — inner lists are same reference! Use comprehension instead
4. **`+=` vs `+`** — `+=` extends in-place, `+` creates new list
5. **Shallow copy** shares nested objects — use `copy.deepcopy()` for full independence
6. **List comprehension if-else** — `[x if cond else y for x in iter]` (if-else BEFORE for)
