## 1. Data Types & Variables
---

```python
# === MUTABLE vs IMMUTABLE ===
# Mutable: list, dict, set, bytearray
# Immutable: int, float, str, tuple, frozenset, bytes, bool

# === FALSY VALUES ===
# 0, 0.0, 0j, "", [], (), {}, set(), None, False, range(0)

# === TYPE CHECKING ===
print(type(42))              # <class 'int'>
print(isinstance(True, int)) # True (bool subclass of int!)
print(isinstance(42, (int, float)))  # True (check multiple types)

# === INTEGER CACHING ===
a = 256; b = 256; print(a is b)   # True  (cached -5 to 256)
a = 257; b = 257; print(a is b)   # False (not cached)

# === FLOAT PRECISION ===
print(0.1 + 0.2 == 0.3)  # False!
import math
print(math.isclose(0.1 + 0.2, 0.3))  # True

# === TRAPS ===
print(bool(""))    # False
print(bool(" "))   # True (space is NOT empty!)
print(type(True + True))  # <class 'int'>, value = 2
```

## 2. Operators
---

```python
# === FLOOR DIVISION & MODULO WITH NEGATIVES ===
print(-7 // 2)   # -4 (floors toward -infinity, NOT toward zero!)
print(-7 % 2)    # 1

# === POWER IS RIGHT-ASSOCIATIVE ===
print(2 ** 3 ** 2)  # 512 = 2 ** 9 (NOT 8 ** 2)

# === LOGICAL OPERATORS RETURN VALUES (not just True/False) ===
print(0 or "" or [] or "hello" or 42)  # "hello" (first truthy)
print(1 and 2 and 3)                     # 3 (last truthy)
print(1 and 0 and 3)                     # 0 (first falsy)
print([] and "hello")                     # [] (first falsy)

# === is vs == ===
# is  -> identity (same object in memory)
# ==  -> equality (same value)
# Always use 'is' with None: x is None

# === WALRUS OPERATOR :=  ===
data = [1, 2, 3, 4, 5]
if (n := len(data)) > 3:
    print(f"Long: {n}")  # Long: 5

# === += TRAP ===
# Mutable:   x += [3] modifies x in-place
# Immutable: x += (3,) creates new tuple
```

## 3. Strings
---

```python
# === IMMUTABLE ===
# s[0] = "X"  -> Error!

# === SLICING BEYOND BOUNDS IS OK ===
print("hello"[1:100])  # "ello" (no error!)

# === KEY METHODS ===
# .find() returns -1 if not found
# .index() raises ValueError if not found
# .split() handles multiple spaces; .split(" ") does NOT

# === STRING FORMATTING ===
name, age = "Alice", 25
print(f"{name} is {age}")        # f-string (Python 3.6+)
print(f"{42:08b}")               # "00101010" (binary with padding)
print(f"{3.14159:.2f}")          # "3.14"
print(f"{'hello':>10}")          # "     hello" (right-aligned)

# === TRAP ===
print("" == False)        # False (different types!)
print(bool("") == False)  # True
```

## 4. Lists
---

```python
# === METHODS THAT RETURN NONE ===
# .sort()    -> sorts in-place, returns None
# .reverse() -> reverses in-place, returns None
# .append()  -> adds item, returns None
# .extend()  -> extends, returns None

# sorted() returns NEW list; .sort() returns None

# === LIST COMPREHENSION ===
# [expr for x in iterable if condition]           # filter
# [expr_if_true if cond else expr_if_false for x]  # transform

# === SHALLOW vs DEEP COPY ===
import copy
a = [[1, 2], [3, 4]]
b = a[:]              # Shallow: inner lists shared!
c = copy.deepcopy(a)  # Deep: fully independent

# === MULTIPLICATION TRAP ===
a = [[0]] * 3      # [[0], [0], [0]] — SAME inner list!
a[0].append(1)
print(a)            # [[0, 1], [0, 1], [0, 1]]

# === += vs + ===
a = [1,2,3]; b = a; b += [4]; print(a)    # [1,2,3,4] (in-place!)
a = [1,2,3]; b = a; b = b + [4]; print(a)  # [1,2,3] (new list!)
```

## 5. Tuples
---

```python
# === SINGLE ELEMENT TRAP ===
print(type((1)))    # int (just parentheses!)
print(type((1,)))   # tuple (comma makes it a tuple)

# === IMMUTABLE BUT... ===
t = (1, [2, 3])
t[1].append(4)      # Works! Mutable element can change
# t[1] += [5]       # TypeError! But list IS modified!

# === ONLY 2 METHODS ===
# .count() and .index()

# === NO TUPLE COMPREHENSION ===
# (x for x in range(5)) is a GENERATOR, not a tuple!

# === CAN BE DICT KEY (if all elements hashable) ===
d = {(1, 2): "point"}  # OK
# d = {(1, [2]): "x"}  # Error! List not hashable
```

## 6. Dictionaries
---

```python
# === KEY MUST BE HASHABLE ===
# OK: int, float, str, tuple (of hashables), frozenset, bool
# NOT OK: list, dict, set

# === True == 1 == 1.0 TRAP ===
d = {True: 'a', 1: 'b', 1.0: 'c'}
print(d)  # {True: 'c'} — all same key!

# === 'in' CHECKS KEYS, NOT VALUES ===
d = {"a": 1, "b": 2}
print("a" in d)  # True
print(1 in d)    # False

# === KEY METHODS ===
# .get(key, default)  — safe access
# .setdefault(key, default)  — add if missing
# .pop(key, default)  — remove and return
# .update(other)  — merge

# === MERGING (Python 3.9+) ===
# d1 | d2  (new dict)
# d1 |= d2 (in-place)

# === fromkeys TRAP ===
d = dict.fromkeys([1,2,3], [])
d[1].append('x')
print(d)  # {1:['x'], 2:['x'], 3:['x']} — SAME list!
```

## 7. Sets
---

```python
# === EMPTY SET ===
s = set()   # Correct
# s = {}    # This is an EMPTY DICT, not a set!

# === KEY OPERATIONS ===
# s1 | s2  (union)        s1 & s2  (intersection)
# s1 - s2  (difference)   s1 ^ s2  (symmetric difference)

# === .remove() vs .discard() ===
# .remove(x) — KeyError if missing
# .discard(x) — no error if missing

# === DEDUPLICATION ===
# list(set([1,1,2,2,3]))        — unordered!
# list(dict.fromkeys([1,1,2,2,3]))  — preserves order!
```

## 8. Control Flow
---

```python
# === for...else / while...else (VERY IMPORTANT!) ===
# else block runs ONLY when loop finishes WITHOUT break
# If break happens, else is SKIPPED

for i in range(5):
    pass
else:
    print("runs!")  # YES — no break happened

for i in range(5):
    if i == 3: break
else:
    print("runs?")  # NO — break happened

# === EMPTY LOOP ELSE ===
for i in range(0):    pass
else: print("yes!")   # YES — empty loop = no break

while False:          pass
else: print("yes!")   # YES — body never ran = no break

# === range(start, stop, step) ===
# stop is NOT included!
# range(5) = 0,1,2,3,4
```

## 9. Functions
---

```python
# === MUTABLE DEFAULT ARGUMENT (CLASSIC TRAP!) ===
def bad(lst=[]):     # DEFAULT LIST IS SHARED!
    lst.append(1)
    return lst

print(bad())  # [1]
print(bad())  # [1, 1] — same list!

def good(lst=None):  # FIX: Use None
    if lst is None: lst = []
    lst.append(1)
    return lst

# === ARGUMENT ORDER ===
# def f(positional, /, normal, *args, keyword_only, **kwargs)

# === SCOPE (LEGB) ===
# Local -> Enclosing -> Global -> Built-in

# === *args = tuple, **kwargs = dict ===
# return a, b  -> returns tuple!
# No return    -> returns None

# === LAMBDA ===
# lambda x, y: x + y  (single expression only)
# sorted(lst, key=lambda x: x[1])

# === KEY HIGHER-ORDER FUNCTIONS ===
# map(func, iterable)    -> applies func to each
# filter(func, iterable) -> keeps truthy results
# reduce(func, iterable) -> cumulative (from functools)
```

## 10. Built-in Functions
---

```python
# === round() BANKER'S ROUNDING ===
print(round(0.5))  # 0 (rounds to even!)
print(round(1.5))  # 2 (rounds to even!)
print(round(2.5))  # 2

# === all() and any() WITH EMPTY ===
print(all([]))  # True  (vacuous truth!)
print(any([]))  # False

# === sorted() RETURNS LIST ===
print(sorted("hello"))  # ['e', 'h', 'l', 'l', 'o'] — LIST!

# === zip() STOPS AT SHORTEST ===
print(list(zip([1,2,3], ['a','b'])))  # [(1,'a'), (2,'b')]

# === enumerate(iterable, start=0) ===
print(list(enumerate('abc', 1)))  # [(1,'a'), (2,'b'), (3,'c')]
```

## 14. Decorators
---

```python
# === BASIC PATTERN ===
from functools import wraps
def decorator(func):
    @wraps(func)  # ALWAYS USE THIS!
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper

# === @wraps preserves __name__, __doc__ ===
# Without it: wrapper.__name__ and wrapper.__doc__ are lost!

# === STACKED DECORATORS: BOTTOM-UP ===
# @a
# @b
# def f(): ...
# = a(b(f))

# === WITH ARGUMENTS: 3 LEVELS ===
# def repeat(n):       <- outer
#     def decorator(f): <- middle
#         def wrapper(): <- inner

# === @lru_cache for memoization ===
```
