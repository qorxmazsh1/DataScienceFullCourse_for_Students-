# 📖 THEORY

```python
# Dict creation
d1 = {"a": 1, "b": 2}
d2 = dict(a=1, b=2)
d3 = dict([("a", 1), ("b", 2)])
d4 = dict.fromkeys(["a", "b", "c"], 0)  # {'a':0, 'b':0, 'c':0}
d5 = {}  # empty dict (NOT empty set!)
print(d1, d2, d3, d4)
```

```python
# Accessing
d = {"a": 1, "b": 2}
print(d["a"])          # 1
# print(d["c"])        # KeyError!
print(d.get("c"))      # None (safe)
print(d.get("c", 0))   # 0 (with default)
```

```python
# Dict methods
d = {"a": 1, "b": 2, "c": 3}
print(d.keys())        # dict_keys(['a', 'b', 'c'])
print(d.values())      # dict_values([1, 2, 3])
print(d.items())       # dict_items([('a',1), ('b',2), ('c',3)])
print(d.pop("a"))      # 1 (removes and returns)
print(d)               # {'b': 2, 'c': 3}
d.setdefault("x", 10)  # Adds 'x':10 if 'x' not present
print(d)

d2 = {"b": 20, "d": 4}
d.update(d2)            # Merges d2 into d
print(d)
```

```python
# Dict comprehension
squares = {x: x**2 for x in range(5)}
print(squares)  # {0:0, 1:1, 2:4, 3:9, 4:16}

# Dict from two lists
keys = ["a", "b", "c"]
vals = [1, 2, 3]
d = dict(zip(keys, vals))
print(d)  # {'a': 1, 'b': 2, 'c': 3}
```

```python
# Python 3.7+: dicts preserve insertion order
d = {"c": 3, "a": 1, "b": 2}
print(list(d.keys()))  # ['c', 'a', 'b'] - insertion order!
```

```python
# Merging dicts
d1 = {"a": 1, "b": 2}
d2 = {"b": 3, "c": 4}

# Method 1: ** unpacking (3.5+)
merged = {**d1, **d2}
print(merged)  # {'a': 1, 'b': 3, 'c': 4}

# Method 2: | operator (3.9+)
merged = d1 | d2
print(merged)  # {'a': 1, 'b': 3, 'c': 4}
```

```python
# 'in' checks KEYS, not values!
d = {"a": 1, "b": 2}
print("a" in d)     # True (key exists)
print(1 in d)       # False (1 is a value, not a key!)
print(1 in d.values())  # True
```

```python
# collections module
from collections import defaultdict, Counter, OrderedDict

# defaultdict
dd = defaultdict(list)
dd["a"].append(1)
dd["a"].append(2)
print(dd)  # defaultdict(<class 'list'>, {'a': [1, 2]})

# Counter
c = Counter("abracadabra")
print(c)                  # Counter({'a': 5, 'b': 2, 'r': 2, 'c': 1, 'd': 1})
print(c.most_common(2))   # [('a', 5), ('b', 2)]
```

# 🔬 DEEP DIVE
---

```python
# TRAP: True/1/1.0 as dict key
d = {True: "a", 1: "b", 1.0: "c"}
print(d)       # {True: 'c'} - all same key!
print(len(d))  # 1
# Because True == 1 == 1.0 and hash(True) == hash(1) == hash(1.0)
```

```python
# TRAP: dict.fromkeys with mutable default
d = dict.fromkeys(["a", "b", "c"], [])
d["a"].append(1)
print(d)  # {'a': [1], 'b': [1], 'c': [1]} - ALL share same list!

# Correct way:
d = {k: [] for k in ["a", "b", "c"]}
d["a"].append(1)
print(d)  # {'a': [1], 'b': [], 'c': []} - independent!
```

```python
# TRAP: List cannot be dict key
d = {}
# d[[1, 2]] = "hello"  # TypeError: unhashable type: 'list'
d[(1, 2)] = "hello"     # OK! Tuple is hashable
print(d)
```

# 🔑 KEY TAKEAWAYS
---
1. **`in` checks keys, not values** — use `in d.values()` for value check
2. **`{True:'a', 1:'b', 1.0:'c'}`** has only 1 key — True==1==1.0
3. **`dict.fromkeys(keys, [])`** shares same mutable default — use comprehension instead
4. **Keys must be hashable** — list can't be key, tuple can
5. **`{}` is empty dict**, not empty set — use `set()` for empty set
6. **Python 3.7+** preserves insertion order
7. **`d[key]` raises KeyError**, `d.get(key)` returns None
