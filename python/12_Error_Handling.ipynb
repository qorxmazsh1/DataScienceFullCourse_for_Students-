```python
# try / except / else / finally
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")       # Error: division by zero
else:
    print("No error occurred")  # Only if NO exception
finally:
    print("Always executes")    # ALWAYS runs
```

```python
# else runs ONLY when no exception
try:
    result = 10 / 2
except ZeroDivisionError:
    print("Error")
else:
    print(f"Result: {result}")  # This prints: Result: 5.0
finally:
    print("Done")
```

```python
# TRAP: finally runs even with return!
def func():
    try:
        return 1
    finally:
        return 2  # This OVERRIDES the return!

print(func())  # 2 (not 1!)
```

```python
# Exception hierarchy (simplified)
# BaseException
#   +-- SystemExit
#   +-- KeyboardInterrupt
#   +-- GeneratorExit
#   +-- Exception
#       +-- ValueError
#       +-- TypeError
#       +-- KeyError
#       +-- IndexError
#       +-- AttributeError
#       +-- NameError
#       +-- StopIteration
#       +-- ArithmeticError -> ZeroDivisionError
#       +-- LookupError -> KeyError, IndexError
#       +-- OSError -> FileNotFoundError
#       +-- ImportError -> ModuleNotFoundError

# Multiple except
try:
    x = int("hello")
except (ValueError, TypeError) as e:
    print(f"Conversion error: {e}")
```

```python
# Custom exceptions
class InsufficientFundsError(Exception):
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Cannot withdraw {amount}, balance is {balance}")

try:
    raise InsufficientFundsError(100, 200)
except InsufficientFundsError as e:
    print(e)  # Cannot withdraw 200, balance is 100
    print(e.balance, e.amount)  # 100 200
```

```python
# raise and re-raise
def validate_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    return age

# Re-raise
try:
    validate_age(-1)
except ValueError:
    print("Logging error...")
    raise  # Re-raises the same exception
```

```python
# EAFP vs LBYL
# LBYL (Look Before You Leap)
d = {"key": "value"}
if "key" in d:
    val = d["key"]

# EAFP (Easier to Ask Forgiveness than Permission) - Pythonic!
try:
    val = d["key"]
except KeyError:
    val = "default"

# Python prefers EAFP style
```

```python
# except Exception vs except BaseException
# except: (bare) catches EVERYTHING including SystemExit, KeyboardInterrupt
# except Exception: catches normal exceptions (recommended)
# except BaseException: same as bare except

# NEVER use bare except: - always catch specific exceptions
```

```python
# TRAP 1: finally overrides return
def tricky():
    try:
        return "try"
    except:
        return "except"
    finally:
        return "finally"

print(tricky())  # "finally" — ALWAYS overrides!
```

```python
# TRAP 2: Exception order matters
try:
    raise ValueError("test")
except Exception:  # Catches first! (parent class)
    print("Exception caught")
except ValueError:  # Never reached!
    print("ValueError caught")
# Better: put specific exceptions FIRST
```

```python
# TRAP 3: Bare except catches too much
# except:   -> catches SystemExit, KeyboardInterrupt too!
# except Exception:  -> recommended! Doesn't catch SystemExit

# TRAP 4: Exception chaining
try:
    try:
        raise ValueError("original")
    except ValueError:
        raise RuntimeError("new") from None  # Suppresses chain
except RuntimeError as e:
    print(e)  # "new" — original is hidden
```

```python
# Solution 3
def retry(func, max_attempts=3):
    last_error = None
    for _ in range(max_attempts):
        try:
            return func()
        except Exception as e:
            last_error = e
    raise last_error

call_count = 0
def failing_func():
    global call_count
    call_count += 1
    if call_count < 3:
        raise ValueError("not yet")
    return "success"

call_count = 0
assert retry(failing_func) == "success"
print("\u2705 Exercise 3 passed!")
```

# 🔑 KEY TAKEAWAYS
---
1. **`else`** runs only when NO exception in try block
2. **`finally`** ALWAYS runs — even with return!
3. **`return` in `finally` overrides** return in try/except
4. **Never use bare `except:`** — use `except Exception:`
5. **EAFP** (try/except) is more Pythonic than LBYL (if checks)
6. **Custom exceptions** should inherit from `Exception`
7. **`assert` is disabled** with `-O` flag
