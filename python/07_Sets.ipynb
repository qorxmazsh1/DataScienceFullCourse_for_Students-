```python
# Set creation
s1 = {1, 2, 3}
s2 = set()           # Empty set (NOT {}!)
s3 = set([1, 2, 2, 3, 3])  # {1, 2, 3} - duplicates removed
print(type({}))      # <class 'dict'> - NOT set!
print(type(set()))   # <class 'set'>
```

```python
# Set properties: unordered, no duplicates, elements must be hashable
s = {3, 1, 4, 1, 5, 9, 2, 6}
print(s)  # Some order (no guaranteed order)
# s = {[1, 2]}  # TypeError! List is not hashable
```

```python
# Set methods
s = {1, 2, 3}
s.add(4)           # {1, 2, 3, 4}
s.remove(2)        # {1, 3, 4} - KeyError if not found!
s.discard(99)      # No error if not found
print(s)

# Set operations
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(a | b)   # Union: {1, 2, 3, 4, 5, 6}
print(a & b)   # Intersection: {3, 4}
print(a - b)   # Difference: {1, 2}
print(a ^ b)   # Symmetric diff: {1, 2, 5, 6}

print(a.issubset(b))     # False
print({3, 4}.issubset(a)) # True
print(a.isdisjoint({7, 8}))  # True (no common elements)
```

```python
# frozenset - immutable set
fs = frozenset([1, 2, 3])
# fs.add(4)   # AttributeError!
d = {fs: "value"}  # Can be dict key!
print(d)
```

```python
# Set comprehension
evens = {x for x in range(10) if x % 2 == 0}
print(evens)  # {0, 2, 4, 6, 8}
```

```python
# Deduplication
lst = [1, 2, 2, 3, 1, 4]
print(list(set(lst)))            # Deduplicated but order may change
print(list(dict.fromkeys(lst)))  # Deduplicated, ORDER PRESERVED
```

# 🔑 KEY TAKEAWAYS
---
1. **`{}` is dict, not set** — use `set()` for empty set
2. **Sets are unordered** — no guaranteed iteration order
3. **Elements must be hashable** — no lists in sets
4. **`remove()` raises KeyError**, `discard()` doesn't
5. **frozenset** is immutable and hashable (can be dict key)
6. **Set operations**: `|` union, `&` intersection, `-` difference, `^` symmetric diff
7. **O(1) membership check** — much faster than list for `in` operator
